# Увод у анализу података

Пред вама је приручник _Увод у анализу података_ у коме у десет лекција дајемо преглед основних вештина које се односе на визуелизацију и обраду података. Идеја приручника је да демонстрира како се користећи програмски језик Пајтон (_Python_) и његове библиотеке уз мало програмерске вештине може из погодно представљених података добити нова вредност -- иноформација. Овај приручник је написан у Џупајтеру (_Jupyter_), што је интерактивно радно окружење за извршавање Пајтон програма.

Језик за кога смо се определили је Пајтон, али циљ овог приручника није учење програмирања у Пајтону. Циљ је усвајање основних вештина и техника визуелизације и обраде података, а програмски језик је само средство које нам омогућује да прецизно формулишемо идеје и да их применимо у пракси. Пајтон је модеран програмски језик који се развија и редовно одржава, и за кога постоји бесплатна подршка. То га чини погодним избором.
